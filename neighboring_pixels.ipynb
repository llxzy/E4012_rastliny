```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data.csv")
df
```

```python
def mean_across_neighbors(row, neighborhood_size=1):
    """
    row - the row of the dataframe to get values for
    neighborhood size - determines the distance in which to check neighbors
    """
    tray_id = row.Tray
    tray_rows = df.loc[df["Tray"] == tray_id]
    
    possible_X_coords = range(row["File X"] - neighborhood_size, row["File X"] + neighborhood_size, 1)
    possible_Y_coords = range(row["File Y"] - neighborhood_size, row["File Y"] + neighborhood_size, 1)
    
    neighborhood = tray_rows.loc[tray_rows['File X'].isin(possible_X_coords) & tray_rows['File Y'].isin(possible_Y_coords)]
    
    value_columns = [c for c in neighborhood.columns if c not in ['Tray', 'ROI name', 'File X', 'File Y']]
    
    
    for col in value_columns:
        row[col] = np.mean(neighborhood[col])
    
    return row

```

```python
df_means = df.apply(lambda row: mean_across_neighbors(row, 1), axis= 1)
df_means
```

```python
df_means.to_csv("data_means.csv")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

```python
data = df_means.drop(columns='ROI name')
labels = df_means['ROI name']
labels = labels.astype('category')
X_train, X_test, y_train, y_test = train_test_split(data, labels, shuffle=True, test_size=0.2)
```

```python
classifier = DecisionTreeClassifier(max_depth=30)
classifier.fit(X_train, y_train)
```

```python
predictions = classifier.predict(X_test)
print(classification_report(y_test, predictions))

# somehow remove the tech?
# problem bude skor asi v spracovani dat nez algoritmoch
# or overfitting
```

```python
import sklearn.decomposition as dcp

df_means_cleaned = df_means.drop(columns=["Tray", "File X", "File Y"])

train_set = df_means_cleaned.drop(columns=["ROI name"])
labels = df_means_cleaned["ROI name"]

```

```python
from sklearn.model_selection import KFold
kfold = KFold(3, shuffle=True, random_state=33)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_set, labels, random_state=33, shuffle=True, test_size=0.3)
```

```python
pca = dcp.PCA(0.99)
X_train = pca.fit(X_train).transform(X_train)
#y_train = pca.fit(y_train).transform(y_train)

X_test = pca.fit(X_test).transform(X_test)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_X = scaler.fit_transform(X_train)

dfX = pd.DataFrame(scaled_X)
dfY = pd.DataFrame(y_train)
```

```python
# svm approach with kfold

from sklearn.svm import SVC

svc = SVC(kernel='rbf')
results = []

for i,j in kfold.split(X_train, y_train):
    Xtrain, Xtest = dfX.iloc[i,:], dfX.iloc[j,:]
    ytrain, ytest = dfY.iloc[i], dfY.iloc[j]
    svc.fit(Xtrain, ytrain.values.ravel())
    predictions = svc.predict(Xtest)
    results.append(classification_report(ytest, predictions))
    
for r in results:
    print(r)
    print()
```

```python
# decision tree using values from neighboring pixels
scXtrain, scXtest, scytrain, scytest = train_test_split(train_set, labels, shuffle=True, random_state=33, test_size=0.3)

scaledXtrain = scaler.fit_transform(scXtrain)
scaledXtest = scaler.fit_transform(scXtest)

tree_classifier = DecisionTreeClassifier(ccp_alpha=0.001, min_samples_leaf=3, min_samples_split=7)

tree_classifier.fit(scaledXtrain, scytrain)
preds = tree_classifier.predict(scaledXtest)
print(classification_report(scytest, preds))
```
